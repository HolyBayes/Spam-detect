```python
%pylab inline

import pandas

feature_names = [
    line.strip().split(':')[0] 
    for line in open('spambase.names').readlines()[33:]
]

spam_data = pandas.read_csv('spambase.data', header=None, names=(feature_names + ['spam']))

X, y = spam_data.ix[:, :-1].values, spam_data.ix[:, -1].values

total_amount = len(spam_data)
print 'Total amount: %i'%(total_amount)
spam_amount = sum(spam_data.ix[i, :]['spam'] for i in xrange(total_amount))
print 'Spam part: %f'%(spam_amount/total_amount)
```

```python
# Task2
trainX = spam_data.ix[:3000, :-1]
trainY = spam_data.ix[:3000, -1]
testX = spam_data.ix[3000:, :-1]
testY = spam_data.ix[3000:, -1]
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

dtclf = DecisionTreeClassifier(max_depth=7, criterion='gini').fit(trainX, trainY)
y_pred = dtclf.predict(testX)

import sklearn.metrics as metrics
print metrics.precision_recall_fscore_support(testY, y_pred)
```

Визуализировав данные, я увидел, что в обучающую выборку попали спамовые письма, а в тестовую - неспамовые (сперва идет спам). Поэтому очень разумно бы было перемешать выборку и обучить дерево на перемешанной выборке. Кроме того, сами оценки качества не очень хорошие - на лекциях приводились такие метрики качества, как критерий бустинга, Джини, точный стохастический тест Фишера.

```python
#dummy classifier (constant model)
from sklearn.dummy import DummyClassifier

dclf = DummyClassifier().fit(trainX, trainY)
y_pred = dclf.predict(testX)

import sklearn.metrics as metrics
print metrics.precision_recall_fscore_support(testY, y_pred)
```

Метрики качества костантной модели ужасные (несмотря на неадекватность данных эвристик можно сказать, что чем они меньше - тем соответствующая им модель хуже)

```python
#decision tree on shuffled sample
import random
shuffled_indexes = range(total_amount)
random.shuffle(shuffled_indexes)

trainX = [spam_data.ix[i, :-1] for i in shuffled_indexes[:3000]]
trainY = [spam_data.ix[i, -1] for i in shuffled_indexes[:3000]]
testX = [spam_data.ix[i, :-1] for i in shuffled_indexes[3000:]]
testY = [spam_data.ix[i, -1] for i in shuffled_indexes[3000:]]

sdtclf = DecisionTreeClassifier(max_depth=7, criterion='gini').fit(trainX, trainY)
y_pred = sdtclf.predict(testX)

print metrics.precision_recall_fscore_support(testY, y_pred)
```

Видно теперь, что при смешивании в тестовой выборке оказались как хорошие письма, так и плохие, притом в сравнимой с общим соотношением пропорцией. В тестовой выборке начали попадаться спамовые, и поэтому точность стала отличной от единицы, как было ранее. Это может навести на простой вывод, что данный алгоритм с отличной точностью определяет хорошие письма и ошибается в основном на спамовых.

```python
#feature importance

most_important_features = argsort(sdtclf.feature_importances_)[::-1]
for idx in most_important_features[:8]:
    print '%d %s, importance = %.2f' % (idx, feature_names[idx], sdtclf.feature_importances_[idx])
```

```python
#kNN
from sklearn.neighbors import KNeighborsClassifier
knnclf = KNeighborsClassifier(n_neighbors=10, metric='minkowski')
knnclf.fit(trainX, trainY)
y_pred = map(lambda x: int(round(x)), knnclf.predict(testX))

print metrics.precision_recall_fscore_support(testY, y_pred)
```

Видно, что на данной выборке по всем рассмотренным ранее оценкам качества данный алгоритм работает чуть хуже, чем дерево решений. Оно и понятно - метрические алгоритмы хороши на задачах кластеризации, а не на задачах с логическими закономерностями 

```python
#normalize features
normalized_data = spam_data
for i in xrange(len(feature_names)):  
    mean = sum(spam_data.ix[j, i] for j in xrange(total_amount)) / total_amount
    sigma = (sum((spam_data.ix[j, i] - mean)**2 for j in xrange(total_amount)) / total_amount)**0.5
    for j in xrange(total_amount):
        normalized_data.ix[j, i] = (spam_data.ix[j, i] - mean) / sigma
```

```python
ntrainX = [normalized_data.ix[i, :-1] for i in shuffled_indexes[:3000]]
ntrainY = [normalized_data.ix[i, -1] for i in shuffled_indexes[:3000]]
ntestX = [normalized_data.ix[i, :-1] for i in shuffled_indexes[3000:]]
ntestY = [normalized_data.ix[i, -1] for i in shuffled_indexes[3000:]]

knnclf1 = KNeighborsClassifier(n_neighbors=10, metric='minkowski')
knnclf1.fit(ntrainX, ntrainY)
y_pred = map(lambda x: int(round(x)), knnclf1.predict(ntestX))

print metrics.precision_recall_fscore_support(ntestY, y_pred)
```

Метрики стали заметно лучше и сравнимы с деревом решений!

Я решил использовать kFold (k = 5). Объяснил это тем, что k = 5 это универсальное значение для этого метода, обычно работающее хорошо (Обычно по дефолту берут k=10, но есть опасения, что тогда в обучающую выборку могут попать только хорошие письма. Поэтому я решил взять чуть больше 3 (спам:хорошие = 1:2) и не экспериментировать). LeaveOneOut здесь неразумно использовать, поскольку данных много.

```python
# Decision Tree (on shuffled data)
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score

cv = KFold(3000, n_folds=5)

depth_values = range(3, 7)  + range(5, 20, 5)
best_depth = 0
best_roc = 0.5
for depth in depth_values:
    clf = DecisionTreeClassifier(max_depth=depth)
    roc = cross_val_score(clf, trainX, trainY, cv=cv, scoring='roc_auc').mean()
    if roc > best_roc:
        best_roc = roc
        best_depth = depth
print 'best depth: %i' % (best_depth)

leaf_values = range(1, 8) + range(5, 20, 5)
best_samples_leaf = 1
best_roc= 0.5
for leaf in leaf_values:
    clf = DecisionTreeClassifier(max_depth=best_depth, min_samples_leaf=leaf)
    roc = cross_val_score(clf, trainX, trainY, cv=cv, scoring='roc_auc').mean()
    if roc > best_roc:
        best_roc = roc
        best_samples_leaf = leaf
print 'best min_samples_leaf: %i' % (best_samples_leaf)

clf = DecisionTreeClassifier(max_depth=best_depth, min_samples_leaf=best_samples_leaf)
clf.fit(trainX, trainY)
probas_clf = clf.predict_proba(testX)
fpr, tpr, thresholds = roc_curve(testY, probas_clf[:, 1])
roc_auc = auc(fpr, tpr)
print 'roc auc: %0.2f' % (roc_auc)
```

ROC AUC остался тем же. Вроде бы ничего не улучшилось. Однако не стоит забывать, что теперь наше дерево гораздо менее переобучено (благодаря кросс-валидации) и на других тестах данный классификатор будет работать лучше. По сути мы избавились от переобучения, ничем при этом не заплатив (остальные два параметра классификатора подбираются аналогично, просто я не нашел эти параметры в мануале на офсайте sklearn)
